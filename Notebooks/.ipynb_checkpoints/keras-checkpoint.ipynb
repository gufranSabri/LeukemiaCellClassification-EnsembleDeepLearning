```python
data_src_path = "C:\\Users\\gufra\\Desktop\\Work\\Projects\\AI\\ALL_Classification\\Datasets\\"
graph_path = r'C:\Users\gufra\Desktop\Work\Projects\AI\ALL_Classification\Models\Graphs90'

TRAIN_DIR = os.path.join(data_src_path, "Split90", "train")
TEST_DIR = os.path.join(data_src_path, "Split90", "test")
VALID_DIR = os.path.join(data_src_path, "Split90", "val")
```

```python
labels = ["all", "hem"]

train_batches = ImageDataGenerator().flow_from_directory(directory=TRAIN_DIR, target_size=(224,224), seed=111, classes=labels, batch_size = 8)
test_batches = ImageDataGenerator().flow_from_directory(directory=TEST_DIR, target_size=(224,224), seed=111, classes=labels, shuffle=False, batch_size = 8)
valid_batches = ImageDataGenerator().flow_from_directory(directory=VALID_DIR, target_size=(224,224), seed=111, classes=labels, batch_size = 8)
```

```python
mbase = tf.keras.applications.MobileNetV2(input_shape=(224, 224, 3),include_top=False, weights='imagenet')
mbase.trainable = False
```

```python
model1 = tf.keras.Sequential()
model1.add(tf.keras.layers.Lambda(tf.keras.applications.mobilenet_v2.preprocess_input, input_shape=(224, 224, 3)))
model1.add(mbase)
model1.add(tf.keras.layers.Flatten())
model1.add(tf.keras.layers.Dense(256, activation='relu'))
model1.add(tf.keras.layers.Dense(2, activation='softmax'))

model1.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history1 = model1.fit(
    x=train_batches,
    steps_per_epoch=len(train_batches),
    validation_data=valid_batches,
    validation_steps=len(valid_batches),
    epochs=80,
    verbose=1,
    callbacks = [
        tf.keras.callbacks.EarlyStopping(
            monitor="val_accuracy",
            verbose=1,
            min_delta=0.001,
            patience=10,
            mode="auto",
            restore_best_weights=True
        )
    ]
)
```

```python
y_pred = model1.predict(test_batches, verbose=1)
y_pred_bool = np.argmax(y_pred, axis=1)

print(classification_report(test_batches.classes, y_pred_bool, digits=4))
```

```python
acc = history1.history['accuracy']
val_acc = history1.history['val_accuracy']
epochs = range(1, len(acc) + 1)
plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.legend()
plt.savefig(os.path.join(graph_path, 'mobilenetv2_acc.png'))
```

## Custom Model

```python
class Structural(tf.keras.layers.Layer):
    def __init__(self):
        super(Structural, self).__init__()

    def get_points_straight(self, img, line_coord, limit, dir):
        first, last=None,None
        for i in range(limit):
            if dir == 1:
                if img[line_coord][i] == 0:
                    if first==None:first = [line_coord, i]
                    else: last = [line_coord, i]
            else:
                if img[i][line_coord] == 0:
                    if first==None:first= [i, line_coord]
                    else: last = [i, line_coord]
        return [first,last]

    def diag_traverser(self, img, x, y, get_points, dir):
        first, last = None, None
        count = 0
        
        i,j=x,y
        while True:
            try:
                if i<0 or j<0:break
                if img[i,j] == 0:
                    if first==None:first= [j,i]
                    else: last = [j,i]
                    count+=1

                i+=1
                if dir==1: j+=1
                else: j-=1
            except:
                break

        if get_points:return [first,last]
        return count

    def get_distance(self, point1, point2):
        if point1==None or point2==None:return -1
        return math.dist(point1,point2)

    def get_distance_between_centers(self, line1_points, line2_points):
        point11, point12 = line1_points[0], line1_points[1]
        point21, point22 = line2_points[0], line2_points[1]

        if None in [point11, point12, point21, point22]: return -1

        line1_center = [(point11[0] + point12[0])/2, (point11[1] + point12[1])/2]
        line2_center = [(point21[0] + point22[0])/2, (point21[1] + point22[1])/2]

        return self.get_distance(line1_center, line2_center)
        
    def call(self, inputs):
        batch = []
        for inp in inputs:
            img = np.uint8(tf.identity(inp).numpy())
            
            bw = cv2.Canny(img, 100, 200)
            bw = cv2.bitwise_not(bw)
            
            h,w = bw.shape
            cropped = bw[:,40:w-40]
            h,w = cropped.shape
            
            partition_offset_x = w//4
            partition_offset_y = h//4
            line1_x = partition_offset_x
            line2_x = partition_offset_x*2
            line3_x = partition_offset_x*3
            line1_y = partition_offset_y
            line2_y = partition_offset_y*2
            line3_y = partition_offset_y*3
            
            points_hor1 = self.get_points_straight(cropped, line1_y, w, 1)
            points_hor2 = self.get_points_straight(cropped, line2_y, w, 1)
            points_hor3 = self.get_points_straight(cropped, line3_y, w, 1)
            points_ver1 = self.get_points_straight(cropped, line1_x, h, 2)
            points_ver2 = self.get_points_straight(cropped, line2_x, h, 2)
            points_ver3 = self.get_points_straight(cropped, line3_x, h, 2)
            points_diag11 = self.diag_traverser(cropped, 0, 0, True, 1)
            points_diag12 = self.diag_traverser(cropped, 0, w//2, True, 1)
            points_diag13 = self.diag_traverser(cropped, h//2, 0, True, 1)
            points_diag21 = self.diag_traverser(cropped, 0,w-1, True, 2)
            points_diag22 = self.diag_traverser(cropped, 0,w//2, True, 2)
            points_diag23 = self.diag_traverser(cropped, h//2,w-1, True, 2)

            disth1 = self.get_distance(points_hor1[0],points_hor1[1])
            disth2 = self.get_distance(points_hor2[0],points_hor2[1])
            disth3 = self.get_distance(points_hor3[0],points_hor3[1])
            distv1 = self.get_distance(points_ver1[0],points_ver1[1])
            distv2 = self.get_distance(points_ver2[0],points_ver2[1])
            distv3 = self.get_distance(points_ver3[0],points_ver3[1])
            distd11 = self.get_distance(points_diag11[0],points_diag11[1])
            distd12 = self.get_distance(points_diag12[0],points_diag12[1])
            distd13 = self.get_distance(points_diag13[0],points_diag13[1])
            distd21 = self.get_distance(points_diag21[0],points_diag21[1])
            distd22 = self.get_distance(points_diag22[0],points_diag22[1])
            distd23 = self.get_distance(points_diag23[0],points_diag23[1])

            cbp_11 = self.diag_traverser(cropped, 0, 0, False, 1)
            cbp_12 = self.diag_traverser(cropped, 0, w//2, False, 1)
            cbp_13 = self.diag_traverser(cropped, h//2, 0, False, 1)
            cbp_21 = self.diag_traverser(cropped, 0, w-1, False, 2)
            cbp_22 = self.diag_traverser(cropped, 0, w//2, False, 2)
            cbp_23 = self.diag_traverser(cropped, h//2, w-1, False, 2)

            block_counts = []
            for i in range(4):
                for j in range(4):
                    block_cropped = cropped[partition_offset_x*i: partition_offset_x*(i+1), partition_offset_y*j: partition_offset_y*(j+1)]
                    block_counts.append(np.count_nonzero(block_cropped==0))

            distc1 = self.get_distance_between_centers(points_hor1, points_ver3)
            distc2 = self.get_distance_between_centers(points_hor2, points_ver2)
            distc3 = self.get_distance_between_centers(points_hor3, points_ver1)
            distc4 = self.get_distance_between_centers(points_diag11, points_diag21)
            distc5 = self.get_distance_between_centers(points_diag12, points_diag23)
            distc6 = self.get_distance_between_centers(points_diag13, points_diag22)

            count = np.count_nonzero(img==0)

            data = []
            data.append(points_hor1[0][1] if (points_hor1[0] is not None) else -1)
            data.append(points_hor2[0][1] if (points_hor2[0] is not None) else -1)
            data.append(points_hor3[0][1] if (points_hor3[0] is not None) else -1)
            data.append(points_ver1[0][0] if (points_ver1[0] is not None) else -1)
            data.append(points_ver2[0][0] if (points_ver2[0] is not None) else -1)
            data.append(points_ver3[0][0] if (points_ver3[0] is not None) else -1)
            data.append(points_hor1[1][1] if (points_hor1[1] is not None) else -1)
            data.append(points_hor2[1][1] if (points_hor2[1] is not None) else -1)
            data.append(points_hor3[1][1] if (points_hor3[1] is not None) else -1)
            data.append(points_ver1[1][0] if (points_ver1[1] is not None) else -1)
            data.append(points_ver2[1][0] if (points_ver2[1] is not None) else -1)
            data.append(points_ver3[1][0] if (points_ver3[1] is not None) else -1) #12

            data.append(points_diag11[0][0] if (points_diag11[0] is not None) else -1)
            data.append(points_diag12[0][0] if (points_diag12[0] is not None) else -1)
            data.append(points_diag13[0][0] if (points_diag13[0] is not None) else -1)
            data.append(points_diag21[0][1] if (points_diag21[0] is not None) else -1)
            data.append(points_diag22[0][1] if (points_diag22[0] is not None) else -1)
            data.append(points_diag23[0][1] if (points_diag23[0] is not None) else -1)
            data.append(points_diag11[1][0] if (points_diag11[1] is not None) else -1)
            data.append(points_diag12[1][0] if (points_diag12[1] is not None) else -1)
            data.append(points_diag13[1][0] if (points_diag13[1] is not None) else -1)
            data.append(points_diag21[1][1] if (points_diag21[1] is not None) else -1)
            data.append(points_diag22[1][1] if (points_diag22[1] is not None) else -1)
            data.append(points_diag23[1][1] if (points_diag23[1] is not None) else -1) #24

            data.append(disth1)
            data.append(disth2)
            data.append(disth3)
            data.append(distv1)
            data.append(distv2)
            data.append(distv3) #30

            data.append(distd11)
            data.append(distd12)
            data.append(distd13)
            data.append(distd21)
            data.append(distd22)
            data.append(distd23) #36

            data.append(cbp_11)
            data.append(cbp_12)
            data.append(cbp_13)
            data.append(cbp_21)
            data.append(cbp_22)
            data.append(cbp_23) #42

            data = data + block_counts #58

            data.append(distc1)
            data.append(distc2)
            data.append(distc3)
            data.append(distc4)
            data.append(distc5)
            data.append(distc6) #64

            data.append(count) #65
            
            batch.append(tf.constant(data))
        
        return tf.constant(np.stack(batch))

class ALL_Net(tf.keras.Model):
    def __init__(self, pretrained_base, 
                 include_pretrained_base, include_structural_features, include_raw_features, 
                 resize_shape=(28,28), **kwargs):
        
        super().__init__(**kwargs)
        self.include_pretrained_base = include_pretrained_base
        self.include_structural_features = include_structural_features
        self.include_raw_features = include_raw_features
        self.resize_shape = resize_shape
        
        self.structural = Structural()
        self.pretrained_base = pretrained_base
        self.flatten = tf.keras.layers.Flatten()
        self.dropout = tf.keras.layers.Dropout(0.2)
        self.fc1 = tf.keras.layers.Dense(256, activation='selu', kernel_initializer="lecun_normal")
        self.fc2 = tf.keras.layers.Dense(2, activation='softmax')
        self.concatenate = tf.keras.layers.Concatenate()
        
    def call(self, inputs):
        final_features = []
        if self.include_pretrained_base:
            pbase_logits = self.pretrained_base(inputs)
            flattened_pbase = self.flatten(pbase_logits)
            
            final_features.append(flattened_pbase)

        if self.include_structural_features:
            final_features.append(self.structural(inputs))
            
        if self.include_raw_features:
            final_features.append(self.flatten(inputs))
            
        concatenated_final_features = self.concatenate(final_features)
        
        fc1_logits = self.fc1(concatenated_final_features)
        fc2_logits = self.fc2(fc1_logits)
        
        return fc2_logits

mobilenetv2 = tf.keras.models.Model(
    inputs=mbase.input, 
    outputs=tf.keras.layers.GlobalAveragePooling2D()(mbase.output)
)

model_allnet = ALL_Net(mobilenetv2, True, True, True)
model_allnet.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'], run_eagerly=True)
```
